# tICA: Try Many Lags

```python
from IPython.parallel import Client
from itertools import repeat
c = Client()
lbv = c.load_balanced_view()
lbv.block = True
```

```python
def ptica(args):
    lt, seqs = args
    from mixtape.tica import tICA
    tica = tICA(n_components=10, lag_time=lt, weighted_transform=True).fit(seqs)
    ts = tica.timescales_
    del tica, seqs
    return lt, ts
```

```python
print(list(range(1,100,5)))
tresults = dict(
    lbv.map(ptica, zip(range(1,100,5), repeat(sa.seqs2d)))
)
```

# Moving Average Trajectory

```python
import pandas as pd
```

```python
txydf = pd.DataFrame(txy)
```

```python
plt.subplots(figsize=(8,5))
stride = 100
ma = pd.rolling_mean(txydf, 4000).dropna()
_ = plt.hist2d(txy[:,0], txy[:,1], bins=90, cmin=1, cmap=plt.get_cmap('terrain'))
plt.plot(txy[::stride, 0], txy[::stride, 1], 'k-', zorder=10, linewidth=1, label='Strided')
plt.plot(ma[::1][0], ma[::1][1], 'r-', zorder=11, linewidth=4, label='Smoothed')
plt.scatter(ma.head(1)[0], ma.head(1)[1], marker='*', zorder=20, c='yellow', s=1000)
plt.legend(loc='lower right')
plt.xlabel('Solvent tIC 1')
plt.ylabel('Solvent tIC 2')
plt.savefig('figs-2-tica/hist_with_traj-lt10.pdf')
```
