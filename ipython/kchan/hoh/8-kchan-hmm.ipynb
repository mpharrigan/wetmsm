# HMM

```python
from mixtape.ghmm import GaussianFusionHMM
```

```python
dim = 2
lt = 10
hmm = GaussianFusionHMM(n_states=8, n_features=dim)
```

```python
hmm.fit([txy[::lt, 0:dim]])
```

## Populations

```python
plt.subplots(figsize=(10,5))
_ = plt.hist2d(txy[:,0], txy[:,1], bins=90, cmin=1, cmap=plt.get_cmap('terrain'))
plt.scatter(hmm.means_[:,0], hmm.means_[:,1], s=500, zorder=10, c=hmm.populations_, norm=plt.Normalize(vmin=0),
            cmap=plt.get_cmap('terrain'), linewidth=2)
plt.title('HMM: dim = %d; lt = %d' % (dim, lt) )
plt.xlabel('Solvent tIC 1')
plt.ylabel('Solvent tIC 2')
plt.colorbar(label='Populations')
plt.savefig('figs-8-hmm/populations-d%d-lt%d.pdf' % (dim, lt))
```

## Timescales

```python
plt.subplots(figsize=(3,5))
plt.hlines(tica.timescales_, 0, 1, 'b', label='tICA')
plt.hlines(hmm.timescales_, 1, 2, 'r', label='HMM')
plt.xlim((0,2))
plt.xticks([])
plt.ylabel('Timescale')
plt.title('Water Timescales (dim = %d)' % dim)
plt.yscale('log')
plt.legend(loc='lower right')
plt.savefig('figs-8-hmm/timescales-d%d.pdf' % dim)
```

## Network

```python
truncmat = np.copy(hmm.transmat_)
truncmat[truncmat < 1e-8] = 0.0
```

```python
plt.imshow(np.log(truncmat), interpolation='none')
plt.colorbar()
```

```python
import networkx as nx
```

```python
weights = np.log(truncmat)
minny = np.min(weights[np.isfinite(weights)])
weights = weights - minny
weights[np.isinf(weights)] = 0
weights
plt.imshow(weights, interpolation='none')
plt.colorbar()
```

```python
import scipy.sparse
```

```python
spweights = scipy.sparse.coo_matrix(weights)
```

```python
plt.subplots(figsize=(10,5))
# Histogram
_ = plt.hist2d(txy[:,0], txy[:,1], bins=90, cmin=1, cmap=plt.get_cmap('terrain'))

# Points
plt.scatter(hmm.means_[:,0], hmm.means_[:,1], s=500, zorder=10, c=hmm.populations_, norm=plt.Normalize(vmin=0),
            cmap=plt.get_cmap('terrain'), linewidth=2)

# Arrows
for r, c, dat in zip(spweights.row, spweights.col, spweights.data):
    if r != c:
       
        plt.annotate('', (hmm.means_[c,0], hmm.means_[c,1]), xytext=(hmm.means_[r,0], hmm.means_[r,1]),
                     arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2', lw=dat / 2.0))

# Titles an stuff
plt.title('HMM: dim = %d; lt = %d' % (dim, lt) )
plt.xlabel('Solvent tIC 1')
plt.ylabel('Solvent tIC 2')
plt.colorbar(label='Populations')
plt.savefig('figs-8-hmm/arrows.pdf')
```

## Eigenvectors

```python
from mixtape.markovstatemodel import MarkovStateModel
```

```python
msm = MarkovStateModel()
msm.transmat_ = np.copy(hmm.transmat_)
```

```python
ev = msm.left_eigenvectors_
ts = np.append([np.inf], msm.timescales_)
```

```python
def plot_eigen(eigeni):
    plt.subplots(figsize=(8,5))
    e1 = ev[:,eigeni] / np.sqrt(np.dot(ev[:,eigeni], ev[:,eigeni]))
    pos_ind = np.where(e1>=0)[0]
    neg_ind = np.where(e1<0)[0]
    print(len(pos_ind), len(neg_ind))
    
    
    plt.scatter(txy[:,0], txy[:,1], s=1, linewidth=0, alpha=0.2, c='g')
    plt.scatter(hmm.means_[pos_ind,0], hmm.means_[pos_ind,1],
                c='r',  linewidth=1.5, s=1000*e1[pos_ind], alpha=0.8, zorder=10)
    plt.scatter(hmm.means_[neg_ind,0], hmm.means_[neg_ind,1],
                c='b',  linewidth=1.5, s=-1000*e1[neg_ind], alpha=0.8, zorder=10)
    plt.title('Eigenvector {}, Timescale {:.1f}'.format(eigeni, ts[eigeni]))
    plt.xlabel('Water tIC 1')
    plt.ylabel('Water tIC 2')
    plt.gca().set_axis_bgcolor('#F9F9F9')
```

```python
plot_eigen(0)
```

```python
plot_eigen(1)
```

```python
plot_eigen(2)
```

```python
plot_eigen(3)
```
