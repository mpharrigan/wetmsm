# MSM

```python
from mixtape.cluster import MiniBatchKMeans
from mixtape.markovstatemodel import MarkovStateModel
```

```python
n_clusters = np.sqrt(len(txy)/2)
n_clusters = int(n_clusters)
print(n_clusters)
```

```python
kmeans = MiniBatchKMeans(n_clusters=n_clusters)
kmeans.fit([tx[:,0:2] for tx in ticax])
```

```python
plt.subplots(figsize=(8,5))
_ = plt.hist2d(txy[:,0], txy[:,1], bins=90, cmin=1, cmap=plt.get_cmap('terrain'))
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=50, zorder=10, c='r')
plt.title('Clustering: dim = 2')
plt.xlabel('Solvent tIC 1')
plt.ylabel('Solvent tIC 2')
plt.savefig('figs-6-msm/cluster-2d.pdf')
```

# Timescales

```python
def ptimescales(args):
    lt, labels = args
    from mixtape.markovstatemodel import MarkovStateModel
    msm = MarkovStateModel(lt, ergodic_cutoff=1).fit(labels)
    return lt, msm.timescales_
```

```python
from IPython.parallel import Client
from itertools import repeat
c = Client()
lbv = c.load_balanced_view()
lbv.block = True
```

```python
timescales = dict(lbv.map(ptimescales, zip(range(1,40,2), repeat(kmeans.labels_))))
```

```python
plt.subplots(figsize=(8,5))
for lt, ts in timescales.items():
    plt.scatter(np.ones(len(ts)) * lt, ts, s=50)
plt.axvline(20)
plt.yscale('log')
plt.xlabel('Lagtime')
plt.ylabel('Implied Timescale')
plt.title('Ergodic cutoff = 1, dim = 2')
plt.savefig('figs-6-msm/it-erg1-2d.pdf')
```

```python
msm = MarkovStateModel(lag_time=20, ergodic_cutoff=1).fit(kmeans.labels_)
```

```python
ev = msm.left_eigenvectors_
ts = np.append([np.inf], msm.timescales_)
```

```python
plt.subplots(figsize=(3,5))
plt.hlines(tica.timescales_, 0, 1, 'b')
plt.hlines(msm.timescales_, 1, 2, 'r')
plt.xlim((0,2))
plt.xticks([])
plt.ylabel('Timescale')
plt.title('Water MSM Timescales')
plt.yscale('log')
plt.savefig('figs-6-msm/timescales-erg1-2d.pdf')
```

# Eigen

```python
def plot_eigen(eigeni):
    plt.subplots(figsize=(8,5))
    e1 = ev[:,eigeni] / np.sqrt(np.dot(ev[:,eigeni], ev[:,eigeni]))
    pos_ind = np.where(e1>=0)[0]
    neg_ind = np.where(e1<0)[0]
    print(len(pos_ind), len(neg_ind))
    
    
    plt.scatter(txy[:,0], txy[:,1], s=1, linewidth=0, alpha=0.2, c='g')
    plt.scatter(kmeans.cluster_centers_[pos_ind,0], kmeans.cluster_centers_[pos_ind,1],
                c='r',  linewidth=1.5, s=1000*e1[pos_ind], alpha=0.8, zorder=10)
    plt.scatter(kmeans.cluster_centers_[neg_ind,0], kmeans.cluster_centers_[neg_ind,1],
                c='b',  linewidth=1.5, s=-1000*e1[neg_ind], alpha=0.8, zorder=10)
    plt.title('Eigenvector {}, Timescale {:.1f}'.format(eigeni, ts[eigeni]))
    plt.xlabel('Water tIC 1')
    plt.ylabel('Water tIC 2')
    plt.gca().set_axis_bgcolor('#F9F9F9')
```

```python
plot_eigen(0)
```

```python
plot_eigen(1)
```

```python
plot_eigen(2)
```

# PCCA

```python
from msmbuilder.lumping import PCCAPlus
import scipy.sparse
```

```python
pcca = PCCAPlus(scipy.sparse.csr_matrix(msm.transmat_), num_macrostates=6)
```

```python
plt.subplots(figsize=(8,5))
_ = plt.hist2d(txy[:,0], txy[:,1], bins=90, cmin=1, cmap=plt.get_cmap('Greys'))
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=70, zorder=10, c=pcca.microstate_mapping)
plt.title('PCCA+')
plt.xlabel('Solvent tIC 1')
plt.ylabel('Solvent tIC 2')
plt.savefig('figs-6-msm/pccaplus-2d.pdf')
```

# Pull Structures

```python
cents = []
for i in range(pcca.num_macrostates):
    whr = np.where(pcca.microstate_mapping == i)[0]
    cents += [np.mean(kmeans.cluster_centers_[whr, 0:2], axis=0)]
cents = np.asarray(cents)
```

```python
plt.subplots(figsize=(8,5))
_ = plt.hist2d(txy[:,0], txy[:,1], bins=90, cmin=1, cmap=plt.get_cmap('Greys'))
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=70, zorder=10, c=pcca.microstate_mapping)
plt.scatter(cents[:,0], cents[:,1], s=200, zorder=20, c='w', linewidth=2)
plt.title('PCCA+')
plt.xlabel('Solvent tIC 1')
plt.ylabel('Solvent tIC 2')
#plt.savefig('figs-6-msm/pccaplus-2d.pdf')
```

```python
# Was going to just consider first traj, too restrictive
tf_stride = 100
tfirst = txy[::tf_stride, 0:2]
```

```python
diff = cents[:, np.newaxis, :]  - np.repeat(tfirst[np.newaxis, ...], 6, axis=0)
print(diff.shape)
```

```python
dotty = np.linalg.norm(diff, axis=2)
```

```python
# Find closest
srted = np.argsort(dotty)
topsrted = srted[:, :10] * tf_stride
```

```python
plt.subplots(figsize=(8,5))
_ = plt.hist2d(txy[:,0], txy[:,1], bins=90, cmin=1, cmap=plt.get_cmap('Greys'))
#plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=70, zorder=10, c=pcca.microstate_mapping)
#plt.scatter(cents[:,0], cents[:,1], s=600, zorder=20, c='#EEEEEE', linewidth=2)

for i, srt in enumerate(topsrted):
    plt.scatter(txy[srt, 0], txy[srt, 1], c=np.ones(len(srt))*i, s=50, edgecolor='k', zorder=30, norm=plt.Normalize(0,6))

plt.title('Pulling Structures')
plt.xlabel('Solvent tIC 1')
plt.ylabel('Solvent tIC 2')
plt.savefig('figs-6-msm/pull_structs-fromstride.pdf')
```
