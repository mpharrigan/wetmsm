```python
drid = verboseload('drid.pickl')
```

```python
dridc = np.concatenate(drid)
# Get rid of first 100 frames. They look to be constrained
dridc = dridc[100:]
```

```python
dridc.shape
```

```python
with open('../hoh/2a-tica.pickl', 'rb') as f:
    wtica = pickle.load(f)
with open('../hoh/2b-ticax.pickl', 'rb') as f:
    wticax = pickle.load(f)
wtxy = wticax[0]
```

# tICA

```python
from mixtape.tica import tICA
```

```python
tica = tICA(n_components=10, lag_time=10, weighted_transform=True)
```

```python
tica.fit([dridc])
```

```python
ticax = tica.transform([dridc])
```

```python
txy = ticax[0]
```

```python
plt.subplots(figsize=(3,5))
plt.hlines(tica.timescales_, 0, 1, 'b', label='DRID')
plt.hlines(wtica.timescales_, 1, 2, 'r', label='Water')
plt.xticks([])
plt.xlim((0, 2))
plt.ylabel('Timescale')
plt.title('DRID tICA Timescales')
plt.yscale('log')
plt.legend(loc='lower right')
plt.savefig('figs-2-tica/timescales.pdf')
```

```python
plt.subplots(figsize=(8,5))
_ = plt.hist2d(txy[:,0], txy[:,1], bins=80, cmin=1)
plt.xlabel('DRID tIC 1')
plt.ylabel('DRID tIC 2')
plt.savefig('figs-2-tica/hist.pdf')
```

```python
import pandas as pd
txydf = pd.DataFrame(txy)
```

```python
plt.subplots(figsize=(8,5))
stride = 100
ma = pd.rolling_mean(txydf, 4000).dropna()
_ = plt.hist2d(txy[:,0], txy[:,1], bins=90, cmin=1, cmap=plt.get_cmap('terrain'))
plt.plot(txy[::stride, 0], txy[::stride, 1], 'k-', zorder=10, linewidth=1, label='Strided')
plt.plot(ma[::1][0], ma[::1][1], 'r-', zorder=11, linewidth=4, label='Smoothed')
plt.scatter(ma.head(1)[0], ma.head(1)[1], marker='*', zorder=20, c='yellow', s=1000)
plt.legend(loc='lower right')
plt.xlabel('DRID tIC 1')
plt.ylabel('DRID tIC 2')
plt.savefig('figs-2-tica/with_path.pdf')
```
