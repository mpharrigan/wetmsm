# Compare

```python
from mixtape.ghmm import GaussianFusionHMM
```

```python
dim = (2, 2)
lt = 10
hmm = GaussianFusionHMM(n_states=10, n_features=np.sum(dim))
```

```python
stacked = np.hstack([cxy[::lt,0:dim[0]], wxy[::lt,0:dim[1]]])
hmm.fit([stacked])
```

```python
plt.subplots(figsize=(8,5))
_ = plt.hist2d(cxy[:,0], wxy[:,0], bins=80, cmin=1, cmap=plt.get_cmap('terrain'))
plt.scatter(hmm.means_[:,0], hmm.means_[:,dim[0]], s=300, c='r', zorder=10)
plt.xlabel('DRID tIC 1')
plt.ylabel('Water tIC 1')
plt.title('HMM: dim = %s' % str(dim))
#plt.savefig('figs-3-hmm/centers.pdf')
```

# Network

```python
weights = np.log(hmm.transmat_)
minny = np.min(weights[np.isfinite(weights)])
weights = weights - minny
weights[np.isinf(weights)] = 0
weights
plt.imshow(weights, interpolation='none')
plt.colorbar()
```

```python
weights = np.copy(hmm.transmat_)
weights[np.arange(len(weights)), np.arange(len(weights))] = 0.0
weights = 100 * weights / np.max(weights)
plt.imshow(weights, interpolation='none')
plt.colorbar()
```

```python
import scipy.sparse
spweights = scipy.sparse.coo_matrix(weights)
```

```python
plt.subplots(figsize=(10,5))
_ = plt.hist2d(cxy[:,0], wxy[:,0], bins=80, cmin=1, cmap=plt.get_cmap('terrain'))
plt.scatter(hmm.means_[:,0], hmm.means_[:,dim[0]], s=300, c=hmm.populations_, zorder=10,
            cmap=plt.get_cmap('terrain'), linewidth=2)


# Arrows
for r, c, dat in zip(spweights.row, spweights.col, spweights.data):
    if r != c:
       
        plt.annotate('', (hmm.means_[c,0], hmm.means_[c,dim[0]]), xytext=(hmm.means_[r,0], hmm.means_[r,dim[0]]),
                     arrowprops=dict(arrowstyle='-|>', connectionstyle='arc3,rad=.2', lw=dat / 10.0))

# Titles an stuff
plt.xlabel('DRID tIC 1')
plt.ylabel('Water tIC 1')
plt.colorbar(label='Populations')
plt.title('HMM: dim = %s' % str(dim))
plt.savefig('figs-3-hmm/arrows.pdf')
```

# Pull Structures

```python
stacked2 = np.hstack([cxy[::100,0:dim[0]], wxy[::100,0:dim[1]]])
cpairs = hmm.draw_samples([stacked2], n_samples=20)
```

```python
plt.subplots(figsize=(8,5))
_ = plt.hist2d(cxy[:,0], wxy[:,0], bins=80, cmin=1, cmap=plt.get_cmap('Greys'))

for i, pairlist in enumerate(cpairs):
    plt.scatter(cxy[pairlist[:,1]*100, 0], wxy[pairlist[:,1]*100,0],
                c=np.ones(len(pairlist)) * i, norm=plt.Normalize(0, cpairs.shape[0]), s=50)
    
    plt.annotate(str(i), hmm.means_[i, [0, dim[0]]], zorder=30, fontsize=14,
                 bbox=dict(boxstyle='round', fc='w', alpha=0.9))

plt.title('HMM: dim = %s' % str(dim) )
plt.xlabel('DRID tIC 1')
plt.ylabel('Water tIC 1')
plt.savefig('figs-3-hmm/pull_structures.pdf')
```
