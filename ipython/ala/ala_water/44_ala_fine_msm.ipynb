# MSM

```python
from mixtape.markovstatemodel import MarkovStateModel
from mixtape.cluster import MiniBatchKMeans
```

```python
kmeans = MiniBatchKMeans(n_clusters=200)
```

```python
kmeans.fit(sa.ticax)
```

```python
plt.subplots(figsize=(8,5))
for txy in sa.ticax:
    plt.scatter(txy[:,0], txy[:,1], alpha=0.1, s=1, linewidth=0)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=50, c='r', zorder=10)
```

```python
msm = MarkovStateModel(lag_time=5)
```

```python
msm.fit(kmeans.labels_)
```

```python
ev = msm.left_eigenvectors_[:,1]

plt.subplots(figsize=(8,5))
for txy in sa.ticax:
    plt.scatter(txy[:,0], txy[:,1], alpha=0.1, s=1, linewidth=0, c='g')
plt.scatter(kmeans.cluster_centers_[ev > 0,0], kmeans.cluster_centers_[ev > 0,1], s=5000*ev[ev >0], c='r', zorder=10)
plt.scatter(kmeans.cluster_centers_[ev < 0,0], kmeans.cluster_centers_[ev < 0,1], s=-5000*ev[ev<0], c='b', zorder=10)
```

```python
# saved every 20 fs
plt.subplots(figsize=(3,5))
plt.hlines(msm.timescales_ * 0.2, 0, 1, 'b')
plt.ylabel('picoseconds')
```

```python
lagtimes = np.arange(1, 40, 1)
timescales = {}
for lt in lagtimes:
    msm = MarkovStateModel(lag_time=lt)
    msm.fit(kmeans2.labels_)
    timescales[lt] = msm.timescales_
```
