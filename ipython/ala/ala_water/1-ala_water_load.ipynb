```python
import tables
```

```python
n_shells = 5
shell_w = 0.2
code = "{:d}x{:d}a".format(n_shells, int(10*shell_w))

seqs = []
fns = []
for i in range(1,8):
    fno = 'trajs/traj-{}.dcd.o.{}.count.h5'.format(i, code)
    fnh12 = 'trajs/traj-{}.dcd.h12.{}.count.h5'.format(i, code)
    
    #fnoa = 'trajs/traj-{}.dcd.o.{}.assign.h5'.format(i, code)
    #fnh12a = 'trajs/traj-{}.dcd.h12.{}.assign.h5'.format(i, code)
    
    fo = tables.open_file(fno)
    fh12 = tables.open_file(fnh12)
    
    #foa = tables.open_file(fnoa)
    #fh12a = tables.open_file(fnh12a)
    
    fp3d_tmp = np.hstack((fo.root.shell_counts, fh12.root.shell_counts))
    print(fp3d_tmp.shape)
    
    seqs.append(fp3d_tmp)
    fns.append(fno)
```

```python
from wetmsm.analysis import SolventShellsAnalysis
```

```python
sa = SolventShellsAnalysis(seqs, shell_w)
```

```python
fp2d = sa.seqs2d[np.random.randint(len(sa.seqs2d))]
for i in np.random.randint(0, fp2d.shape[1], 4):
    plt.plot(fp2d[::100, i], '.-', label="Feature {}".format(i))
plt.xlabel('Time'); plt.ylabel('Shells FP');
plt.legend(loc='best')
```
