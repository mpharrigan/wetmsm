```python
import pickle
```

```python
with open('../drid/drid_tica_transform.pickl', 'rb') as f:
    dseq = pickle.load(f)
with open('../hoh/fp_ic.pickl', 'rb') as f:
    wseq = pickle.load(f)
```

```python
# conf, hoh
dim = (2, 2)
lt = 5
tseq = [
    np.hstack((cs[::lt,:dim[0]], ws[::lt,:dim[1]]))
    for cs, ws in zip(dseq, wseq)
]
```

```python
wxys = np.concatenate(wseq)
dxys = np.concatenate(dseq)
```

```python
plt.subplots(figsize=(8,5))
_ = plt.hist2d(wxys[:,0], dxys[:,1], bins=80, cmin=10)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax)

plt.xlabel('Water tIC 1')
plt.ylabel('DRID tIC 2')
```

# HMM

```python
from mixtape.ghmm import GaussianFusionHMM
```

```python
hmm = GaussianFusionHMM(n_states=8, n_features=np.sum(dim))
```

```python
hmm.fit(tseq)
```

```python
plt.subplots(figsize=(8,5))
_ = plt.hist2d(wxys[:,0], dxys[:,1], bins=80, cmin=1, cmap=plt.get_cmap('terrain'))

plt.scatter(hmm.means_[:,dim[0]+0], hmm.means_[:,1], s=100, c='r')

plt.xlabel('Water tIC 1')
plt.ylabel('DRID tIC 2')
```

```python
plt.subplots(figsize=(8,5))
_ = plt.hist2d(wxys[:,0], dxys[:,0], bins=80, cmin=1, cmap=plt.get_cmap('terrain'))

plt.scatter(hmm.means_[:,dim[0]+0], hmm.means_[:,0], s=500, c='r')

plt.xlabel('Water tIC 1')
plt.ylabel('DRID tIC 1')
```

```python
plt.subplots(figsize=(8,5))
_ = plt.hist2d(dxys[:,0], dxys[:,1], bins=80, cmin=1, cmap=plt.get_cmap('terrain'))

plt.scatter(hmm.means_[:,0], hmm.means_[:,1], s=100, c='r')

plt.xlabel('DRID tIC 1')
plt.ylabel('DRID tIC 2')
```

```python
plt.subplots(figsize=(3,5))
plt.hlines(hmm.timescales_, 0, 1, 'b')
plt.yscale('log')
```

# Arrows

```python
weights = np.copy(hmm.transmat_)
weights[np.arange(len(weights)), np.arange(len(weights))] = 0.0
weights = 100 * weights / np.max(weights)
plt.imshow(weights, interpolation='none')
plt.colorbar()
```

```python
import scipy.sparse
spweights = scipy.sparse.coo_matrix(weights)
```

```python
plt.subplots(figsize=(10,5))
_ = plt.hist2d(wxys[:,0], dxys[:,1], bins=80, cmin=1, cmap=plt.get_cmap('terrain'))

plt.scatter(hmm.means_[:,dim[0]], hmm.means_[:,1], s=500, c=hmm.populations_,
            cmap=plt.get_cmap('Purples'), lw=2)

plt.xlabel('Water tIC 1')
plt.ylabel('DRID tIC 2')
plt.colorbar(label='Populations')


# Arrows
for r, c, dat in zip(spweights.row, spweights.col, spweights.data):
    if r != c:
        plt.annotate('', (hmm.means_[c,dim[0]], hmm.means_[c,1]), xytext=(hmm.means_[r,dim[0]], hmm.means_[r,1]),
                     arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2', lw=dat / 10.0))
        
plt.savefig('figs-3-hmm/arrows-waterdrid.pdf')
```
