```python
_shape_check = [ts.shape for ts in tseq]
print(_shape_check)
```

```python
n_datapoint = int(np.sum([len(ts) for ts in tseq]))
n_cluster = int(np.ceil(np.sqrt(n_datapoint/2)))
print("{} datapoints. Using {} clusters".format(n_datapoint, n_cluster))
```

```python
kmeans = KMeans(n_clusters=n_cluster, n_jobs=6)
```

```python
kmeans.fit(tseq)
```

```python
plt.subplots(figsize=(8,5))
for ts in tseq:
    plt.scatter(ts[:,0], ts[:,hmn_c], s=1, linewidth=0, alpha=0.2, c='g')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,hmn_c], facecolors='none')
plt.xlabel('Conf. tIC 1')
plt.ylabel('Water tIc 1')
```

```python
plt.subplots(figsize=(8,5))
for ts in tseq:
    plt.scatter(ts[:,hmn_c], ts[:,hmn_c+1], s=1, linewidth=0, alpha=0.2, c='g')
plt.scatter(kmeans.cluster_centers_[:,hmn_c], kmeans.cluster_centers_[:,hmn_c+1], facecolors='none')
plt.xlabel('Water tIC 1')
plt.ylabel('Water tIc 2')
```

```python
plt.subplots(figsize=(8,5))
for ts in tseq:
    plt.scatter(ts[:,0], ts[:,1], s=1, linewidth=0, alpha=0.2, c='g')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], facecolors='none')
plt.xlabel('Conf. tIC 1')
plt.ylabel('Conf. tIc 2')
```

```python
lagtimes = list()
timescales = list()
for lt in range(1, 30, 2):
    print(lt)
    msm = MarkovStateModel(lag_time=lt, n_timescales=20)
    msm.fit(kmeans.labels_)
    print("trim", np.sum([v==-1 for v in msm.mapping_.values()]))
    timescales.append(msm.timescales_)
    lagtimes.append([lt] * msm.n_timescales)
```

```python
timescales = np.array(timescales).flatten()
lagtimes = np.array(lagtimes).flatten()
plt.subplots(figsize=(8,5))
plt.scatter(lagtimes, timescales, c='b')
plt.yscale('log')
```

```python
msm = MarkovStateModel(lag_time=11, n_timescales=20)
msm.fit(kmeans.labels_)
```

```python
plt.hlines(msm.timescales_, 0, 1, 'b')
plt.yscale('log')
plt.title('Timescales at lt = {}'.format(msm.lag_time))
```

```python
ev = msm.eigenvectors_
ts = np.append([np.inf], msm.timescales_)
def plot_eigen(eigeni):
    plt.subplots(figsize=(8,5))
    e1 = ev[:,eigeni] / np.sqrt(np.dot(ev[:,eigeni], ev[:,eigeni]))
    pos_ind = np.where(e1>=0)[0]
    neg_ind = np.where(e1<0)[0]
    print(len(pos_ind), len(neg_ind))
    
    for fp in tseq:
        plt.scatter(fp[:,0], fp[:,hmn_c], s=1, linewidth=0, alpha=0.2, c='g')
    plt.scatter(kmeans.cluster_centers_[pos_ind,0], kmeans.cluster_centers_[pos_ind,hmn_c],
                c='r',  linewidth=1.5, s=1000*e1[pos_ind], alpha=0.8, zorder=10)
    plt.scatter(kmeans.cluster_centers_[neg_ind,0], kmeans.cluster_centers_[neg_ind,hmn_c],
                c='b',  linewidth=1.5, s=-1000*e1[neg_ind], alpha=0.8, zorder=10)
    plt.title('Eigenvector {}, Timescale {:.1f}'.format(eigeni, ts[eigeni]))
    plt.xlabel('Conformational tIC 1')
    plt.ylabel('Conformational tIC 2')
    plt.gca().set_axis_bgcolor('#F9F9F9')
    #plt.savefig("dridmsm_eigenv-{}".format(eigeni))
```

```python
plot_eigen(0)
```

```python
plot_eigen(1)
```
