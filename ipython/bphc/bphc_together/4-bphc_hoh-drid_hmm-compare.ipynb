# HMM

```python
from mixtape.ghmm import GaussianFusionHMM
```

```python
dim = (4, 4)
lt = 5
```

```python
dhmm = GaussianFusionHMM(n_states=3, n_features=dim[0])
dhmm.fit([ds[::lt,:dim[0]] for ds in dseq])
```

```python
whmm = GaussianFusionHMM(n_states=3, n_features=dim[1])
whmm.fit([ws[::lt,:dim[1]] for ws in wseq])
```

```python
hmm = GaussianFusionHMM(n_states=3, n_features=np.sum(dim))
hmm.fit([np.hstack([ds[::lt,:dim[0]], ws[::lt,:dim[1]]]) for ds, ws in zip(dseq, wseq)])
```

```python
plt.subplots(figsize=(3,5))

plt.hlines(whmm.timescales_, 0, 1, 'b', lw=2, label='Solvent')
plt.hlines(dhmm.timescales_, 1, 2, 'r', lw=2, label='Conf.')
plt.hlines(hmm.timescales_,  2, 3, 'purple', lw=2, label='Combined')

plt.legend(loc='lower left')

plt.xticks([])
plt.xlim((0,3))
plt.yscale('log')
plt.ylabel('Timescales')
```
