```python
import pickle
from mixtape.cluster import MiniBatchKMeans
from mixtape.markovstatemodel import MarkovStateModel
```

```python
with open('../drid/drid_tica_transform.pickl', 'rb') as f:
    dseq = pickle.load(f)
```

```python
with open('../fp_ic.pickl', 'rb') as f:
    wseq = pickle.load(f)
```

```python
hmn_c = 1
hmn_w = 1
tseq = [
    np.hstack((cs[:,:hmn_c], ws[:,:hmn_w]))
    for cs, ws in zip(dseq, wseq)
]
```

```python
plt.subplots(figsize=(8,5))
for s1, s2 in zip(wseq, dseq):
    plt.scatter(s1[:,0], s2[:,0], s=1, linewidth=0, alpha=0.1, c='b')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax)
plt.plot(x, x, c='r', zorder=20, linewidth=2)
plt.xlim((xmin, xmax))
plt.xlabel('Water tIC 1')
plt.ylabel('DRID tIC 1')
plt.savefig('drid-vs-water.png')
```

```python
wxys = np.concatenate(wseq)
dxys = np.concatenate(dseq)
```

```python
plt.subplots(figsize=(8,5))
_ = plt.hist2d(wxys[:,0], dxys[:,0], bins=80, cmin=10)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax)
plt.plot(x, x, c='r', zorder=20, linewidth=2)
plt.xlim((xmin, xmax))

plt.xlabel('Water tIC 1')
plt.ylabel('DRID tIC 1')
plt.savefig('drid-vs-water.hist.png')
```

# MSM

```python
kmeans = MiniBatchKMeans(n_clusters=150)
```

```python
kmeans.fit(tseq)
```

```python
plt.subplots(figsize=(8,5))
for s1, s2 in zip(wseq, dseq):
    plt.scatter(s1[:,0], s2[:,0], s=1, linewidth=0, alpha=0.1, c='b')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,hmn_c], s=40, c='r')
plt.title('Clustering')
plt.savefig('dridwater2x2.clustering.png')
```

```python
lagtimes = np.arange(1,30,2)
ltdict = {}
for lt in lagtimes:
    msm = MarkovStateModel(lag_time=lt, trim_weight=4)
    msm.fit(kmeans.labels_)
    ltdict[lt] = msm.timescales_
```

```python
for k, v in ltdict.items():
    plt.scatter(k * np.ones(len(v)), v)
plt.yscale('log')
```

```python
msm = MarkovStateModel(lag_time=10)
msm.fit(kmeans.labels_)
```
