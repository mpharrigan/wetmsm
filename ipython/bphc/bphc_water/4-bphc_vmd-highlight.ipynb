### Logistic

```python
def logi(x, mu=0.8, s=0.1):
    return 1.0 / (1 + np.exp(-(x-mu)/s))
```

```python
x = np.linspace(0,5)
plt.plot(x, logi(x, 0.8, 0.1))
```

```python
plt.subplots(figsize=(8,5))
xx = np.arange(len(stic1))
plt.scatter(xx[stic1>0.1], logi(stic1[stic1>0.1]), c='r', s=50, linewidth=1)
plt.xlabel('Feature')
plt.ylabel('Loading')
plt.savefig('loading.png')
#plt.scatter(range(len(stic2)), logi(stic2), c='b', s=10, linewidth=0)
```
