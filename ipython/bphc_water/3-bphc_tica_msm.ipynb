# Build an MSM from tICA

```python
_, n_features = fp_ic[0].shape
n_trajs = len(fp_ic)
n_datapoint = int(np.sum([len(fp2d) for fp2d in fp_ic]))
n_cluster = int(np.ceil(np.sqrt(n_datapoint/2)))
print("{} features in {} trajs\nwith {} total data points\ngiving {} \
datapoints per traj\nor {:.2f} datapoints per feature\nUsing {} clusters".format(
            n_features, n_trajs, n_datapoint, n_datapoint/n_trajs, n_datapoint / (n_features), n_cluster))
```

```python
from mixtape.cluster import KMeans
from mixtape.markovstatemodel import MarkovStateModel
```

```python
kmeans = KMeans(n_clusters=n_cluster, n_jobs=6)
```

```python
# Cluster over spec. number of features (tics)
n_feat = 2
kmeans.fit([fp[:,:n_feat] for fp in fp_ic])
```

```python
for fp in fp_ic:
    plt.scatter(fp[:,0], fp[:,1], s=1, linewidth=0, alpha=0.2)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='r')
```

```python
sym = 'transpose'
lagtimes = list()
timescales = list()
for lt in range(1, 30, 2):
    print(lt)
    msm = MarkovStateModel(lag_time=lt, n_timescales=20, reversible_type=sym)
    msm.fit(kmeans.labels_)
    timescales.append(msm.timescales_)
    lagtimes.append([lt] * msm.n_timescales)
```

```python
timescales = np.array(timescales).flatten()
lagtimes = np.array(lagtimes).flatten()
print(timescales.shape, lagtimes.shape)
```

```python
plt.scatter(lagtimes, timescales)
plt.yscale('log')
```

## Eigenvectors

```python
ev = msm.eigenvectors_
ts = np.append([np.inf], msm.timescales_)
```

```python
def plot_eigen(eigeni, sym):
    plt.subplots(figsize=(8,5))
    e1 = ev[:,eigeni] / np.sqrt(np.dot(ev[:,eigeni], ev[:,eigeni]))
    pos_ind = np.where(e1>=0)[0]
    neg_ind = np.where(e1<0)[0]
    print(len(pos_ind), len(neg_ind))
    
    for fp in fp_ic:
        plt.scatter(fp[:,0], fp[:,1], s=1, linewidth=0, alpha=0.2, c='g')
    plt.scatter(kmeans.cluster_centers_[pos_ind,0], kmeans.cluster_centers_[pos_ind,1],
                c='r',  linewidth=1.5, s=1000*e1[pos_ind], alpha=0.8, zorder=10)
    plt.scatter(kmeans.cluster_centers_[neg_ind,0], kmeans.cluster_centers_[neg_ind,1],
                c='b',  linewidth=1.5, s=-1000*e1[neg_ind], alpha=0.8, zorder=10)
    plt.title('Eigenvector {}, Timescale {:.1f}'.format(eigeni, ts[eigeni]))
    plt.xlabel('Water tIC 1')
    plt.ylabel('Water tIC 2')
    plt.gca().set_axis_bgcolor('#F9F9F9')
    plt.savefig("watermsm_eigenv-{}-{}".format(sym, eigeni))
```

```python
plot_eigen(0, sym)
```

```python
plot_eigen(1, sym)
```

```python
plot_eigen(2, sym)
```

```python
plot_eigen(3, sym)
```

### Pick out some structures

```python
e1 = ev[:,1] / np.sqrt(np.dot(ev[:,1], ev[:,1]))
```

```python
mine = np.argmin(e1)
maxe = np.argmax(e1)
```

```python
plot_eigen(1)
plt.scatter(kmeans.cluster_centers_[mine,0], kmeans.cluster_centers_[mine,1], c='w', s=200)
plt.scatter(kmeans.cluster_centers_[maxe,0], kmeans.cluster_centers_[maxe,1], c='k', s=200, edgecolor='w')
```

```python
wheremax = [np.where(lab==maxe)[0] for lab in kmeans.labels_]
wheremin = [np.where(lab==mine)[0] for lab in kmeans.labels_]
print("These trajectories have structures:")
print([i for i, wm in enumerate(wheremax) if len(wm)>0])
print([i for i, wm in enumerate(wheremin) if len(wm)>0])
```
