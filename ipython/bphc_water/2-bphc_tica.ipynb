# tICA

```python
from sklearn.decomposition import PCA, KernelPCA
from mixtape.tica import tICA
```

```python
#ica = PCA(n_components=2)
ica = tICA(n_components=10, lag_time=10, weighted_transform=True)
```

### Fit all trajectories

```python
ica.fit(fp2d_all)
```

### Force it to eigensolve

```python
print(ica.components_[0].shape)
print(ica.components_[1].shape)
```

```python
tic1 = ica.components_[0] ** 2
tic2 = ica.components_[1] ** 2
```

```python
# Make the biggest one 1
scale = np.max(tic1)
tic1 /= scale
tic2 /= scale
```

```python
plt.scatter(range(len(tic1)), tic1, c='r', s=10, linewidth=0)
plt.scatter(range(len(tic2)), tic2, c='b', s=10, linewidth=0)
```

## tICA Timescales

```python
plt.subplots(figsize=(3,5))
plt.hlines(ica.timescales_, 0, 1, 'b')
plt.xticks([])
plt.ylabel('Timescale')
plt.title('Water tICA Timescales')
plt.yscale('log')
plt.savefig('water_tica_timescales.png')
```

## 2D Plot

```python
fp_ic = ica.transform(fp2d_all)
```

```python
txy = np.vstack([fp[:,:2] for fp in fp_ic])
```

```python
print(txy.shape)
```

```python
import scipy.stats
```

```python
kde = scipy.stats.gaussian_kde(txy.T)
```

```python
plt.subplots(figsize=(8,5))
npoints = 100
xlin = np.linspace(np.min(txy[:,0]), np.max(txy[:,0]), npoints)
ylin = np.linspace(np.min(txy[:,1]), np.max(txy[:,1]), npoints)
xx, yy = np.meshgrid(xlin, ylin)
pos = np.vstack([xx.ravel(), yy.ravel()])
pos.shape
zz = kde.evaluate(pos)
zz = np.reshape(zz.T, xx.shape)
plt.contourf(xx,yy,zz, 200)
plt.xlabel('Water tIC 1')
plt.ylabel('Water tIC 2')
plt.savefig('water_tic_kde.png')
```

### Get Indices

```python
tot_n = np.sum(len(fp) for fp in fp_ic)
tot_n
```

```python
lind = []
rind = []
for i, fp in enumerate(fp_ic):
    x = fp[:,0]
    y = fp[:,1]
    
    rl = np.sqrt((x-mlx)**2 + (y-mly)**2)
    rr = np.sqrt((x-mrx)**2 + (y-mry)**2)
    
    # Make array
    rar = np.where(rr < r)[0]
    lar = np.where(rl < r)[0]
    
    # Add traj index
    #rar = np.vstack((i * np.ones(len(rar), dtype=int), rar))
    #lar = np.vstack((i * np.ones(len(lar), dtype=int), lar))
    
    rind.append((i, rar))
    lind.append((i, lar))
    
```

```python
linda = np.hstack(lind).T
rinda = np.hstack(rind).T
print(len(rinda), len(linda))
print("{:.1f}% {:.1f}%".format(100*len(rinda)/tot_n, 100*len(linda)/tot_n))
print()

linda = linda[::10]
rinda = rinda[::10]
print(len(rinda), len(linda))
print("{:.1f}% {:.1f}%".format(100*len(rinda)/tot_n, 100*len(linda)/tot_n))
```
