## Load in fingerprint

```python
import tables
import glob
```

```python
fp3d_all = []
fn_all = []
for i in range(1,8):
    fno = 'trajs/shells-traj-{}.dcd_count.h5'.format(i)
    fnh1 = 'trajs/traj-{}.dcd.h1.count.h5'.format(i)
    fnh2 = 'trajs/traj-{}.dcd.h2.count.h5'.format(i)
    
    fo = tables.open_file(fno)
    fh1 = tables.open_file(fnh1)
    fh2 = tables.open_file(fnh2)
    
    fp3d_tmp = np.hstack((fo.root.shell_counts, fh1.root.shell_counts, fh2.root.shell_counts))
    print(fp3d_tmp.shape)
    
    fp3d_all.append(fp3d_tmp)
    fn_all.append(fnh1)
```

# Remove zero-variance features, reshape

```python
import wetmsm.analysis as wa
import logging
wa.log.setLevel(logging.INFO)
```

```python
# Normalize
fp3dn_all = [wa.normalize(fp3d, shell_w=0.2) for fp3d in fp3d_all]
```

```python
# Reshape
fp2dup_all = [wa.reshape(fp3d) for fp3d in fp3dn_all]
```

```python
# Prune
fp2d_all, to_delete = wa.prune_all(fp2dup_all)
print(to_delete)
```
