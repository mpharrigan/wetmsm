## Eigenvectors

```python
ev = msm.eigenvectors_
```

```python
def plot_eigen(eigeni):
    e1 = ev[:,eigeni] / np.sqrt(np.dot(ev[:,eigeni], ev[:,eigeni]))
    pos_ind = np.where(e1>=0)[0]
    neg_ind = np.where(e1<0)[0]
    print(len(pos_ind), len(neg_ind))
    
    for fp in fp_ic:
        plt.scatter(fp[:,0], fp[:,1], s=1, linewidth=0, alpha=0.2)
    plt.scatter(kmeans.cluster_centers_[pos_ind,0], kmeans.cluster_centers_[pos_ind,1], c='r', s=1000*e1[pos_ind])
    plt.scatter(kmeans.cluster_centers_[neg_ind,0], kmeans.cluster_centers_[neg_ind,1], c='y', s=-1000*e1[neg_ind])
```

```python
plot_eigen(1)
```
