## Logistic

```python
def logi(x, mu=0.8, s=0.1):
    return 1.0 / (1 + np.exp(-(x-mu)/s))
```

## Traj

```python
traj = md.load('data/SYS62/0_centered.xtc', top='system_eq1.pdb')
```

```python
solute_ind = np.loadtxt('solute_ca_indices.dat', dtype=int, ndmin=2)
solvent_ind = np.loadtxt('solvent_indices.dat', dtype=int, ndmin=2)
```

## Translate

```python
to3d = {}
to2d = {}
absi = 0
pruni = 0
for ute in range(n_solute):
    for sh in range(n_shell):
        if not np.in1d(absi, to_delete):
            to3d[pruni] = (ute, sh)
            to2d[(ute,sh)] = pruni
            pruni += 1
        else:
            to2d[(ute,sh)] = -1
        absi += 1
```

```python
feati = np.argmax(stic1)
featu, feats = to3d[feati]
print(feati)
print(featu, feats)
```

## Save

```python
cd ~/implement/wetmsm/bphc/
```

```python
toinclude = np.where(stic1 > 0.1)[0]
print(toinclude)
```

```python
f = open('trj62_all.dat', 'w')
for fr in range(0, traj.n_frames, 1):
    assn1 = assn[np.where(assn[:,0]==fr)[0], ...]
    towrite = np.zeros(traj.n_atoms)
    
    # Loop over features
    for feati in toinclude:
        featu, feats = to3d[feati]
        rows = np.where(np.logical_and(assn1[:,2]==featu, assn1[:,3]==feats))[0]
    
        highlight = solvent_ind[assn1[rows, 1]]    
        towrite[highlight[:,0]] += stic1[feati]
    
    [f.write('{} '.format(d)) for d in towrite]
    f.write('\n')
    if fr % 100 == 0:
        print(fr)
f.close()
    
```
