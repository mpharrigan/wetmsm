```python
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
```

## Make sample data

```python
centers = [[1,1], [-1,-1], [1,-1]]
X, labels_true = make_blobs(n_samples=750, centers=centers, cluster_std=0.3)
```

```python
plt.subplots(figsize=(8,5))
plt.scatter(X[:,0], X[:,1], s=100, alpha=0.8, c='purple')
plt.title('Data')
plt.savefig('kmeans_data.png')
```

## Do KMeans

```python
cd ~/implement/wetmsm/bphc/
```

```python
k = 3
kmeans = KMeans(n_clusters=k)
kmeans.fit(X)
X1 = kmeans.predict(X)
```

```python
plt.subplots(figsize=(8,5))
plt.scatter(X[:,0], X[:,1], s=100, alpha=0.8, c=X1)
plt.title('KMeans Results')
plt.savefig('kmeans_normal.png')
```
