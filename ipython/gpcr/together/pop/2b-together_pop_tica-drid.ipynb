```python
from scipy import stats
```

```python
plt.subplots(figsize=(8,5))

_ = plt.hist2d(dtxy[:,0], txy[:,0], bins=100, cmin=1, cmap=plt.get_cmap('terrain'), norm=LogNorm())

# 1:1
xx = np.linspace(*plt.xlim())
#plt.plot(xx, xx, 'r--', lw=2, label='y = x')

# Best fit
slope, intercept, r_value, p_value, slope_std_error = stats.linregress(dtxy[:,0], txy[:,0])
plt.plot(xx, slope * xx + intercept, 'r', lw=2, label='Best Fit')

plt.xlabel('DRID tIC 1')
plt.ylabel('POP tIC 1')
plt.legend(loc='lower right')
plt.savefig('figs-2-tica/popdrid-1.pdf')
```
