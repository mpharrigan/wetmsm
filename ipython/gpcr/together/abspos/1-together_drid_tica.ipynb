# Sort

```python
with open('1-2rh1_fns.pickl', 'rb') as f:
    ifns = pickle.load(f)
with open('1-3p0g_fns.pickl', 'rb') as f:
    afns = pickle.load(f)
```

```python
def get_num(fns):
    return np.asarray([int(re.match('.*clone([0-9]+).*', fn).group(1)) for fn in fns])
```

```python
ifn2 = get_num(ifns)
afn2 = get_num(afns)
```

```python
isorty = sorted(zip(iseqs, ifns, ifn2), key=lambda xx: xx[-1])
asorty = sorted(zip(aseqs, afns, afn2), key=lambda xx: xx[-1])
```

```python
seqs = [seq for seq, _, _ in isorty] + [seq for seq, _, _ in asorty]
```

```python
for s in seqs:
    assert s.shape[1] == 825
```

# Do tICA

```python
tica = tICA(n_components=10, weighted_transform=True)
```

```python
tica.fit(seqs)
```

```python
tseqs = tica.transform(seqs)
txy = np.concatenate(tseqs)
```

```python
# Also keep seperate
tiseqs = tica.transform(iseqs)
tixy = np.concatenate(tiseqs)

taseqs = tica.transform(aseqs)
taxy = np.concatenate(taseqs)
```
