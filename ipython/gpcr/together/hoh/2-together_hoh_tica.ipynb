```python
from mixtape.cluster import MiniBatchKMeans
from mixtape.markovstatemodel import MarkovStateModel
```

```python
kmeans = MiniBatchKMeans(n_clusters=100)
```

```python
ttseqs = [seq[:,0:3] for seq in tseqs]
```

```python
kmeans.fit(ttseqs)
```

```python
_ = plt.hist2d(txy[:,0], txy[:,1], bins=100, cmin=1)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='r', zorder=10)
```

```python
msm = MarkovStateModel(lag_time=1).fit(kmeans.labels_)
```
