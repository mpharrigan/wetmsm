```python
seqs = []
fns = []
for count_fn in glob.iglob('3p0g/generation2/trr/*/clone*.trr.pop.count.h5'):
    count_f = tables.open_file(count_fn)
    count_a = np.array(count_f.root.shell_counts)
    
    if len(count_a) > 1:
        seqs.append(count_a)
        fns.append(count_fn)
    count_f.close()
    
```
