# Make a dict fns

```python
with open('abspos/1-seqs.fns.pickl', 'rb') as f:
    afns = pickle.load(f)
with open('1-pop-fns.pickl', 'rb') as f:
    wfns = pickle.load(f)
```

```python
adict = {}
for aseq, afn in zip(aseqs, afns):
    clone_name = afn.split('/')[-1].split('.')[0]
    adict[clone_name] = aseq
```

```python
wdict = {}
for wseq, wfn in zip(wseqs, wfns):
    clone_name = wfn.split('/')[-1].split('.')[0]
    wdict[clone_name] = wseq
```

## In one but not the other

```python
set(adict.keys()) ^ set(wdict.keys())
```

## Take intersection

```python
clones = set(adict.keys()) & set(wdict.keys())
```

```python
set(adict.keys()) - clones
```

```python
set(wdict.keys()) - clones
```

# Plot
