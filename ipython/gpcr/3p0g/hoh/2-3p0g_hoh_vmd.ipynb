# VMD

```python
from wetmsm.vmd_write import VMDWriter, VMDSCRIPT
from os.path import join as pjoin
```

### Load Assignments File

```python
biox_prefix = '/home/harrigan/biox/implement/water/gpcr/3p0g/generation2'
assn_fn = pjoin(biox_prefix, 'trr/8/clone8645.trr.hoh.assign.h5')
assn_f = tables.open_file(assn_fn)
assn = assn_f.root.assignments
print(assn.shape)
```

### Load Indices

```python
ca_inds = np.loadtxt(pjoin(biox_prefix, 'ca_3p0g.dat'), dtype=int)
pop_inds = np.loadtxt(pjoin(biox_prefix, 'hoh_3p0g.dat'), dtype=int)
print(ca_inds.shape, pop_inds.shape)
```

### Load topology

```python
import mdtraj as md
top = md.load('vis/3p0g.pdb')
print(top)
```

### Visualize loadings

```python
tic1 = tica.components_[0]
xx = np.arange(len(tic1))
```

```python
plt.subplots(2,2, figsize=(10,6))
plt.subplot(2,2,1)
plt.scatter(xx,tic1, linewidth=0, s=30)
plt.subplot(2,2,2)
tic1s = tic1 ** 2
tic1s /= np.max(tic1s)
plt.scatter(xx, tic1s, linewidth=0, s=30, c='r')
loading1 = np.copy(tic1s)
loading1[tic1s < 0.1] = 0.0
plt.subplot(2,2,3)
plt.scatter(xx, loading1, linewidth=0, s=30, c='purple')
```

### Make object and compute

```python
vmd = VMDWriter(assn=assn, solvent_ind=pop_inds, n_frames=tseqs[mi].shape[0], n_atoms=top.n_atoms,
                n_solute=len(ca_inds), n_shells=5)
```

```python
user = vmd.compute(loading1, deleted)
```

### Save

```python
np.savetxt('vis/8645.hoh.dat', user, fmt='%.5f')
```

### Generate script

```python
with open('vis/8645.hoh.tcl', 'w') as f:
    f.write(VMDSCRIPT.format(top_fn='3p0g.pdb', traj_fn='clone8645.trr', step=1, dat_fn='8645.hoh.dat'))
```
